```python
test = []

for file_name in file_names: 
    
    if file_name.endswith('.jpg'): 
        
        image_1 = cv2.imread(train_path + file_name)
        image_2 = cv2.imread(train_dotted_path + file_name)
        image_3 = cv2.absdiff(image_1,image_2)

        height,width,channels = image_3.shape 

        mask_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2GRAY)
        mask_2[mask_2 < 20] = 0
        mask_2[mask_2 > 0] = 255

        image_3 = cv2.bitwise_or(image_3, image_3, mask=mask_2)         
        image_3 = cv2.cvtColor(image_3, cv2.COLOR_BGR2GRAY)
        image_4 = np.zeros((height,width,3), np.uint8)

        blobs = skimage.feature.blob_log(image_3, min_sigma=3, max_sigma=4, num_sigma=1, threshold=0.02)

        for blob in blobs:
            y, x, s = blob
            b,g,r = image_2[int(y)][int(x)][:]
            cv2.circle(image_4, (int(x),int(y)), 10, (int(b),int(g),int(r)), thickness=-1) 
            test.append((int(y),int(x)))

        image_4 = cv2.normalize(image_4, image_4, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)

        cv2.imwrite(train_blob_path + file_name, image_4)
        
```
